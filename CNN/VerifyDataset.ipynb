```python

with h5py.File('C:/Users/62811/PCV/Big Project/CNN//DatasetH5PY/datasetCards.hdf5', 'r') as f:

    print(list(f.keys()))    
    x_train = f['x_train']  
    y_train = f['y_train']  

    x_train = np.array(x_train)  
    y_train = np.array(y_train)  


    x_validation = f['x_validation']  
    y_validation = f['y_validation']  

    x_validation = np.array(x_validation)  
    y_validation = np.array(y_validation)  
    
    
    x_test = f['x_test']  
    y_test = f['y_test']  

    x_test = np.array(x_test)
    y_test = np.array(y_test)   
```

```python

%matplotlib inline


plt.rcParams['figure.figsize'] = (10.0, 60.0)

figure, ax = plt.subplots(nrows=25, ncols=4)

for i in range(25):
    for j in range(4):
        ii = np.random.randint(low=0, high=x_train.shape[0])
        ax[i, j].imshow(x_train[ii].astype('uint8'))
        ax[i, j].set_title(labels[y_train[ii]], fontsize=16)
        ax[i, j].axis('off')


plt.tight_layout()
figure.savefig('plot_100_card_images.png')
plt.show()
```

```python

%matplotlib inline

plt.rcParams['figure.figsize'] = (10.0, 7.0)

classesIndexes, classesFrequency = np.unique(y_train, return_counts=True)

print('classes indexes:' , classesIndexes)
print('classes frequency:', classesFrequency)

figure = plt.figure()

plt.bar(classesIndexes, classesFrequency, align='center', alpha=0.6)

plt.ylabel('Jumlah gambar setiap kelas', fontsize=16)

plt.xticks(classesIndexes, fontsize=16)

plt.title('Histogram Cards Dataset', fontsize=20)

figure.savefig('Cardsimages.png')

plt.show()
```
